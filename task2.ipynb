```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1,2,3,4,5 ])
ypoints = np.array([2,4,6,8,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
x_scatter = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y_scatter = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.figure()
plt.scatter(x_scatter, y_scatter, s=100, c='red', alpha=0.9)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

plt.figure()
plt.bar(categories, values, color=['green', 'blue', 'red', 'pink', 'purple'])
plt.title("Bar Plot")
plt.xlabel("Categories" )
plt.ylabel("Values")
plt.show()
```

```python
da= np.random.randn(1000)

plt.figure()
plt.hist(da, bins=40, color='green', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.axis('equal')
plt.show()
```
